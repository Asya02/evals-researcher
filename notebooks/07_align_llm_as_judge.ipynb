```python
import json
```

```python
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv())
```

```python
PHOENIX_COLLECTOR_ENDPOINT = 'http://localhost:6006'
PHOENIX_COLLECTOR_ENDPOINT_FULL='http://localhost:6006/v1/traces'
```

```python
PROJECT_NAME = "gpt-researcher"
```

```python
import phoenix as px

from phoenix.trace import SpanEvaluations
from phoenix.trace.dsl import SpanQuery

import nest_asyncio
nest_asyncio.apply()
```

```python
query = SpanQuery().where(
    "'google search queries to search online' in input.value",
).select(
        input="input.value",
        output="output.value",
)

search_queries_df = px.Client(
    endpoint=PHOENIX_COLLECTOR_ENDPOINT
).query_spans(
    query, 
    project_name=PROJECT_NAME, 
    timeout=None,
    limit=50
)
search_queries_df
```

```python
search_queries_df["input"] = search_queries_df["input"].apply(lambda x: json.loads(x)['messages'][0][0]['kwargs']['content'])
```

```python
search_queries_df["output"] = search_queries_df["output"].apply(lambda x: json.loads(x)["generations"][0][0]["text"])
```

```python
search_queries_df = search_queries_df.reset_index()
```

```python
search_queries_df = search_queries_df.rename(columns={"context.span_id": "id"})
```

```python
search_queries_df.to_csv("../data/annotations/llm_judge_eval_data.csv")
```
