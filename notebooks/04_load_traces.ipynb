## Labeling

```python
import json
```

```python
mask = (spans_df["span_kind"] == "RETRIEVER") & (
    spans_df["attributes.retrieval.documents"].isna() | 
    (spans_df["attributes.retrieval.documents"].apply(lambda x: x == "[]"))
)

no_search_results_df = spans_df.loc[mask]
```

```python
import ast
import json
from pandas import isnull


def is_not_valid_output_message(x):
    if isnull(x) or not isinstance(x, str) or x.strip() == '[]':
        return True
    x = ast.literal_eval(x)
    try:
        if not isinstance(x, list) or len(x) == 0:
            return True
            
        last_msg = x[-1]
        
        if not isinstance(last_msg, dict) or 'message.content' not in last_msg:
            return True
            
        content = last_msg['message.content']
        
        if not isinstance(content, str):
            return True
            
        parsed = json.loads(content)
        
        required_fields = {'server', 'agent_role_prompt'}
        if not all(field in parsed for field in required_fields):
            return True
            
        if not isinstance(parsed['server'], str) or len(parsed['server']) == 0:
            return True
            
        return False
        
    except Exception:
        return True

mask = (spans_df["attributes.llm.input_messages"].notna() & 
    spans_df["attributes.llm.input_messages"].str.contains('You are a seasoned finance analyst AI assistant') &
    spans_df["attributes.llm.output_messages"].apply(is_not_valid_output_message)
    )

not_valid_json_df = spans_df.loc[mask]
```

```python
errors_df = pd.concat([no_search_results_df, not_valid_json_df])
```

```python
annotations = {row['context.span_id']: {"approved": False} for _, row in errors_df.iterrows()}
```

```python
with open('../data/annotations/trace_annotations.json', 'w', encoding='utf-8') as f:
    json.dump(annotations, f, ensure_ascii=False, indent=2)
```
